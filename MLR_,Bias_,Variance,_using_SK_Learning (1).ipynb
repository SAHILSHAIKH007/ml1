```python
import numpy as np
import pandas as pd
# Data visualisation
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
adv=pd.read_csv("advertising.csv") ## Reading the file
adv
```

```python
adv.info()
```

```python
adv.describe()
```

```python
x=adv[['TV','Radio','Newspaper']] ## assigning values to input variable
y=adv['Sales'] ## Value assigned to output variable
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42) ## 80 % Training , 20 % Testing ## Random pattern no is 42
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
mlr=LinearRegression() ## Model is defined
mlr.fit(x_train,y_train) ## Model is trained
```

```python
print("Intercept",mlr.intercept_) # Value of b0
print("Coefficients:") ## Value of b1

list(zip(x,mlr.coef_))
```

```python
mlr_pred=mlr.predict(x_test) ##Prediction of model
mlr_pred
```

```python
Test_mlr_diff = pd.DataFrame({'Actual value':y_test, 'Predicted value': mlr_pred})
Test_mlr_diff
```

```python
mlr_pred_mean=Test_mlr_diff['Predicted value'].mean()
Test_mlr_diff['(mlr_pred_mean-mlr_pred)^2']=(Test_mlr_diff['Predicted value']-mlr_pred_mean )**2
Test_mlr_diff
```

```python
Test_error=((Test_mlr_diff['Actual value'] - Test_mlr_diff['Predicted value'])**2).mean()
Test_error
```

```python
Variance= Test_mlr_diff['(mlr_pred_mean-mlr_pred)^2'].mean()
Variance
```

```python
mlr_train_pred=mlr.predict(x_train) ##Prediction of model
mlr_train_pred
```

```python
Test_mlr_train_diff = pd.DataFrame({'Actual value':y_train, 'Predicted value': mlr_train_pred})
Test_mlr_train_diff
```

```python
mlr_pred_train_mean=Test_mlr_train_diff['Predicted value'].mean()
mlr_pred_train_mean
```

```python
Test_mlr_train_diff['mlr_pred_train_mean-actual value']=abs(mlr_pred_train_mean-Test_mlr_train_diff['Actual value'])
Test_mlr_train_diff
```

```python
Train_error= ((Test_mlr_train_diff['Actual value'] - Test_mlr_train_diff['Predicted value'])**2).mean()
Train_error
```

```python
bias=Test_mlr_train_diff['mlr_pred_train_mean-actual value'].mean()
bias
```
